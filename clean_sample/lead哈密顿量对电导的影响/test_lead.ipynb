```python
%matplotlib tk
import kwant
import numpy as np
import matplotlib.pyplot as plt


Lx=2
m=1
t=2

def make_system():
    #### Define the scattering region. ####
    def chain(pos):
        x = pos[0]
        return 0<=x<=Lx

    def onsite(site):
        return m

    def hop_x(site1,site2):
        x1=site1.pos[0]
        x2=site2.pos[0]
        return t

    sys = kwant.Builder()
    lat=kwant.lattice.chain(a=1,norbs=1,name="chain")
    sys[lat.shape(chain, (0,))] = onsite

    sys[kwant.builder.HoppingKind((1,),lat)]=hop_x


    # left lead
    sym_left_lead = kwant.TranslationalSymmetry((-1,))
    left_lead = kwant.Builder(sym_left_lead)
    left_lead[lat(0,)] = m

    '''这里是lead内的hopping，lead和S区之间的hopping应该不影响散射矩阵'''
    left_lead[lat(1,), lat(0,)] = t
    sys.attach_lead(left_lead)
    
   
    # right lead
    sym_right_lead = kwant.TranslationalSymmetry((1, ))
    right_lead = kwant.Builder(sym_right_lead)
    right_lead[lat(0,)] = m
    right_lead[lat(1, ), lat(0, )] = t
    sys.attach_lead(right_lead)
    return sys


sys=make_system().finalized()
kwant.plot(sys);


sys_ham=sys.hamiltonian_submatrix()
print(sys_ham,'\n')

lead_ham=sys.leads[0].hamiltonian_submatrix()
print(lead_ham,'\n')


# lead1=sys.leads[1].hamiltonian_submatrix()
# print(lead1)
```

```python
# Now that we have the system, we can compute conductance

kwant.plotter.bands(sys.leads[0],momenta=601)
# kwant.plotter.bands(sys.leads[1],momenta=601)

energies = []
data = []

eng=np.linspace(-10,10.1,101)
for i in range(len(eng)):
    # print(eng[i])
    # compute the scattering matrix at a given energy
    smatrix = kwant.smatrix(sys, eng[i])
    print(smatrix.submatrix(0,0).shape)

    # compute the transmission probability from lead 0 to lead1
    energies.append(eng[i])
    # print(smatrix.transmission(0,0))
    data.append(smatrix.transmission(0,1))


# Use matplotlib to write output
# We should see conductance steps
plt.figure()
plt.plot(energies, data)
plt.xlabel("energy [t]")
plt.ylabel("conductance [e^2/h]")
plt.show()
```
