# T=50mK,此代码必须结合Vdis=0中的数据进行修改

$G(E_{i})=\frac{2e^{2}}{h}\int\bar{T(E)}F_{T}(E-E_{i})dE$  \
with $F_{T}(E-E_{i})$ being the thermal broadening function  \
$F_{T}(E-E_{i})=-\frac{\partial}{\partial E}(\frac{1}{exp[(E-E_{i})/k_{B}T+1]})=\frac{1}{4k_{B}T}\frac{1}{cosh^{2}[\frac{E-E_{i}}{2*k_{B}T}]}$

```python
%matplotlib tk
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot
import os


path = '/home/cusps/python/ML_disorder/clean_sample/mu=0/GLR/'

GLR = np.load(path + "GLR.npy")
print(GLR.shape)



#用于积分的E
bias=np.linspace(-0.1,0.1,GLR.shape[1]) 
#新的Ei
bias_T=np.linspace(-0.1,0.1,801)



#KT energy in 50mK
KT= 4.31*1e-3 #meV

#all params are numbers
def dF(E,V,KT):
    return 1/(4*KT)*1/(np.cosh((E-V)/(2.*KT))**2)



#新电导的维度
gLR=np.zeros((len(bias_T),len(GLR[:,0])))

# 矩形积分
for i in range(len(bias_T)):
    for j in range(len(bias)):
        gLR[i,:] += GLR[:,j] * dF(bias[j],bias_T[i],KT) * (bias[2] - bias[1])

print(np.array(gLR).shape)
```

```python
def plot_cond(x,y,cond,title):
    X, Y = np.meshgrid(x, y)
    Z = np.array(cond)
    plt.pcolormesh(X, Y, Z, cmap='bwr', shading='auto') 
    plt.title(title)
    plt.xlabel('B')
    plt.ylabel('bias')
    
    cbar = plt.colorbar()  
    plt.show()


B=np.linspace(0,0.8,GLR.shape[0])

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plot_cond(B,bias,GLR.T/2,'T=0')
plt.subplot(1,2,2)
plot_cond(B,bias_T,gLR/2,'T=50mK')
```

# 处理colorbar




```python
from matplotlib.colors import TwoSlopeNorm
def plot_cond(x, y, G, title):
    X, Y = np.meshgrid(x, y)
    Z = np.array(G)
    
    # 定义 colormap 和归一化
    cmap = 'bwr'  # 蓝-白-红色彩映射
    norm = TwoSlopeNorm(vmin=np.min(Z), vcenter=0, vmax=np.max(Z))  # 零值为白色
    
    # 绘图
    plt.pcolormesh(X, Y, Z, cmap=cmap,shading='auto', vmin=-0.02, vmax=0.02)
    plt.title(title)
    plt.xlabel('B')
    plt.ylabel('bias')
    
    # 添加 colorbar 并标注
    cbar = plt.colorbar()
    cbar.set_label('2e^2/h', fontsize=12)
    
    # 显示图像
    plt.show()

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plot_cond(B,bias,GLR.T/2,'T=0')
plt.subplot(1,2,2)
plot_cond(B,bias_T,gLR/2,'T=50mK')
```
