```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
from scipy import optimize,sparse,interpolate
import kwant
import math

import dolfin
import time
import mshr
from numpy import sqrt,pi,cos,sin,kron
from scipy.sparse.linalg import eigsh
import scipy
# from sys1d import *
from fenics import *


me = 9.109e-31    #kg
hbar = 1.055e-34    #J.S
eV, e = 1.6023e-19, 1.6023e-19
akwant,aFEM = 1e-9,1e-9 #m
meff_n = 0.023*me


Lz=100
hopping_normal = hbar**2/(2*meff_n*akwant**2)/eV   #量纲为eV

#normal region
tx=hopping_normal
ty=tx
tz=tx

eps0 = 8.854e-12  #unit:F/m
eps_r= 15.15

#boundary condition
Vg=-0.1
W=0.1

class def_Parameters(object):
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
class def_Dimension(object):
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
class def_Environment(object):
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)

def kwant_1d_selfconsistant(Parameters_bar,environment_bar,a_lenth):
    Lz = environment_bar.Lz
    akwant=Parameters_bar.akwant
    tx=Parameters_bar.tx
    ty=Parameters_bar.ty
    tz=Parameters_bar.tz

    n=Lz//a_lenth+1
    mu=2*tz*(1+cos(n*pi/(n+1)))
    def sys_shape(pos):
        z=pos[0]   
        return ( 0 <= z <= Lz)  
    def onsite(site,kx,ky,pot):
        z=site.pos[0]
        return 2*tz+tx*kx**2+ty*ky**2+pot[int(z)]-mu
        
    def hop_z(site1,site2):
        z1=site1.pos[0]
        z2=site2.pos[0]
        return -tz
    sys = kwant.Builder()
    lat_n=kwant.lattice.chain(a=a_lenth,norbs=1,name="normal")

    sys[lat_n.shape(sys_shape, (1,))] = onsite
    sys[kwant.builder.HoppingKind((1,),lat_n,lat_n)]=hop_z    
    return sys    

def mesh_1d_sm(environment_bar):
    Lz=environment_bar.Lz   
    mesh = IntervalMesh(1000*Lz, 0, Lz) 
    V = FunctionSpace(mesh,"P",1)
    mesh_points_all = mesh.coordinates()
    mesh_length_sites = len(mesh_points_all)
    return  mesh, V, mesh_length_sites, mesh_points_all


def transform_kwant_to_poisson(kwant_density,Dimension_bar,Parameters_bar):
    site_points = Dimension_bar.site_points
    mesh_points = Dimension_bar.mesh_points
    poisson_scale = Parameters_bar.poisson_scale
    
    poisson_density = scipy.interpolate.griddata(site_points, kwant_density,
                                           mesh_points, method='cubic')

    poisson_density[np.isnan(poisson_density)] = 0
    poisson_density =poisson_scale*np.round(poisson_density, 8)
    return poisson_density

def reorder_source_coordinates(poisson_density,Dimension_bar):
    mesh = Dimension_bar.mesh
    V = Dimension_bar.V
    mesh_points=Dimension_bar.mesh_points
    
    to_1d = lambda x: x.view(dtype=np.dtype([('f1', 'float')]))[:, 0]
    n,d = V.dim(), mesh.geometry().dim()               
    V_dof_coordinates = V.tabulate_dof_coordinates()    
    V_dof_coordinates.resize((n,d))   
    
    ordering = np.argsort(to_1d(V_dof_coordinates))
    wire_indices = np.searchsorted(to_1d(V_dof_coordinates),
                                   to_1d(mesh_points), sorter=ordering)
    values_func = dolfin.Function(V)  
    values_func.vector()[ordering[wire_indices]] = poisson_density[:,0]
    return values_func


def potential_1d_sm(poisson_density_func,Dimension_bar,Parameters_bar,environment_bar):
    V=Dimension_bar.V
    Lz=environment_bar.Lz  
    W=Parameters_bar.W
    Vg=Parameters_bar.Vg
    eps_r=Parameters_bar.eps_r
    u = TrialFunction(V)
    v = TestFunction(V)
    es = 1e-9

    def boundary_up(x):
        return  ( Lz-es  <x[0] < Lz+es )
    def boundary_down(x):
        return  ( 0-es <x[0] < 0+es )

    bc_sc1 = DirichletBC(V, dolfin.Constant(-W-Vg),boundary_up)
    bc_sc2 = DirichletBC(V, dolfin.Constant(-W),boundary_down)

    fs = poisson_density_func        
    fa = dolfin.inner(eps_r*grad(u),grad(v))*dx          
    s = Function(V)

    solve(fa == fs*v*dx, s , [bc_sc1,bc_sc2])    
    poisson_pot = s.vector()[:]
    return poisson_pot


def cal_charge_density(ks,energys,states,sys):
    rho = kwant.operator.Density(sys)
    charge_density = np.zeros(len(sys.sites))
    dks = ks[1]-ks[0]

    for i in range(len(ks)):
        wave_density = np.zeros(len(sys.sites))
        engs=energys[i]
        wavfs=states[i]
    
        for j in range(len(engs)):
            energy=engs[j]
            state=wavfs[:,j]
            if energy <=0:
                wave_density+=rho(state)
        charge_density+=2*wave_density*ks[i]*dks/(2*pi*akwant**3)
    return charge_density

poisson_scale = e*aFEM**2/(eps0)
Parameters_bar = def_Parameters(akwant=akwant,tx=tx,ty=ty,tz=tz, W = W, Vg = Vg,  poisson_scale = poisson_scale,
                                eps0 = eps0, eps_r = eps_r)
environment_bar = def_Environment(Lz = Lz)

#晶格长度
a1=1
system=kwant_1d_selfconsistant(Parameters_bar,environment_bar,a1).finalized()
site_points = np.array([system.sites[i].pos for i in range(len(system.sites))])
mesh,V, mesh_length_sites,mesh_points = mesh_1d_sm(environment_bar)

Dimension_bar = def_Dimension(sys = system, mesh = mesh, V=V,
                                  mesh_length_sites = mesh_length_sites,
                                  mesh_points = mesh_points,
                                  site_points = site_points)
sys = Dimension_bar.sys

print(f"Vg={Vg}","\n")
print(f"W={W}","\n")
print(f"mesh_length_sites={mesh_length_sites}","\n")
print(f"Lz={Lz}nm","\n")
kwant.plot(sys)
```

# self consist

```python
#和其他代码里的逻辑功能不一样
def transform_poisson_to_kwant(poisson_pot,Dimension_bar):
    mesh=Dimension_bar.mesh
    V=Dimension_bar.V
    pot_poisson_to_kwant = dolfin.Function(V)
    pot_poisson_to_kwant.vector()[:] = poisson_pot
    def kwant_pot(*pos):
        return pot_poisson_to_kwant(dolfin.Point(*pos))

    pot=np.zeros(len(site_points))
    for i in range(len(site_points)):
        pot[i]=kwant_pot(i)
    return pot


from fenics import *
t1=time.time()

ks=np.linspace(0,1,1001)*pi

def self_consistent(initial_pot):
    energys=[]
    states=[]
    
    def cal_energy(ks):
        hamiltonian = sys.hamiltonian_submatrix(sparse=False,params = dict(kx=0,ky=ks,pot = initial_pot))
        energy,state = np.linalg.eigh(hamiltonian)
        return energy,state
    
    for i in range(len(ks)):
        energy,state= cal_energy(ks[i])
        energys.append(energy)
        states.append(state)
    
    kwant_density=cal_charge_density(ks,energys,states,sys)
    poisson_density =transform_kwant_to_poisson(kwant_density,Dimension_bar,Parameters_bar)
    poisson_density_func=reorder_source_coordinates(poisson_density,Dimension_bar)  
    poisson_pot_new = potential_1d_sm(poisson_density_func,Dimension_bar,Parameters_bar,environment_bar)
    
    # pot=list(transform_poisson_to_kwant(poisson_pot_new,Dimension_bar))
    pot=transform_poisson_to_kwant(poisson_pot_new,Dimension_bar)

    check = sum(np.abs(pot-initial_pot)**2)/(len(initial_pot))
    print("error：",check)
    
    return pot,check,poisson_density,poisson_pot_new

Error=[]
# pot,check,poisson_density,poisson_pot_new=self_consistent(initial_pot)
# Error.append(check)


t2=time.time()
print("time:",t2-t1)
```

# 梯度下降

```python
def gradient(pot_in):
    # 设置随机种子以确保结果可重复（可选）
    np.random.seed(0)
    # 生成随机数组

    random=1e-9
    random_pot = np.random.uniform(-random, random, len(pot_in))
    #计算梯度
    pot1,_,_,_=self_consistent(pot_in)    
    pot2,_,_,_=self_consistent(pot_in + random_pot)
    
    diff=(pot2-pot1)/random_pot-1

    # print(diff)
    
    grad =2/len(pot1)* (pot1 - pot_in)*diff
    return grad
    
kappa=0.2

Grad=gradient(initial_pot)
pot_next=initial_pot-kappa*Grad


plt.figure()
plt.plot(site_points,initial_pot,label='pot_in')
plt.plot(site_points,pot_next,label='pot_out')
plt.title(f"kappa={kappa}")
plt.xlim(0,100)
plt.legend()
plt.show()
```

```python
NN = 200
for n in range(NN):
    from fenics import *
    t1=time.time()
    print(f"{n+1}th iteration")
    initial_pot = pot_next.copy()

    pot,check,poisson_density,poisson_pot_new=self_consistent(initial_pot)
    Error.append(check)

    Grad=gradient(initial_pot)
    # plt.plot(site_points,Grad)
    pot_next=initial_pot-kappa*Grad
        
    t2=time.time()
    print('time:',t2-t1)
```
